```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap
import plotly.express as px
```

```python
df = pd.read_csv("usgs_main.csv")

print("Data sample:")
df.head()
```

```python
print("Columns:", df.columns)
```

```python
df['tmp_id'] = range(1, len(df) + 1)
lat_lon_cols = ['latitude', 'longitude', 'tmp_id']
other_cols = [c for c in df.columns if c not in ['latitude', 'longitude']]

df_latlon = df[lat_lon_cols].copy()
df_other = df[other_cols].copy()
```

```python
df_latlon
```

```python
df_other
```

```python
df_merged = pd.merge(df_other, df_latlon, on='tmp_id', how='left')

df_merged.drop(columns=['tmp_id'], inplace=True)
```

```python
df_merged
```

```python
for col in df.select_dtypes(include='object').columns:
    df[f'{col}_has_number'] = df[col].str.contains(r'\d', na=False)
    df[f'{col}_lower'] = df[col].str.lower()
```

```python
df
```

```python
print("Columns:", df.columns)

lat_col = "latitude"
lon_col = "longitude"
```

```python
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df[lon_col], df[lat_col]), crs="EPSG:4326"
)
```

```python
gdf.head()
```

```python
map_explore = gdf.explore(
    column=None, cmap="Blues", tiles="CartoDB positron",
    tooltip=df.columns.tolist(), marker_kwds={"radius": 5}
)

map_explore
```

```python
fig_plotly = px.density_mapbox(
    df, lat=lat_col, lon=lon_col,
    radius=10,
    center={"lat": df[lat_col].mean(), "lon": df[lon_col].mean()},
    zoom=5, mapbox_style="carto-positron"
)

fig_plotly.show()
```

```python
m = folium.Map(location=[df[lat_col].mean(), df[lon_col].mean()], zoom_start=5)
heat_data = df[[lat_col, lon_col]].values.tolist()
HeatMap(heat_data, radius=12, blur=15, max_zoom=13).add_to(m)

m
```

```python
import pydeck as pdk

point_layer = pdk.Layer(
    "ScatterplotLayer",
    data=df,
    get_position='[longitude, latitude]',  # استبدل بالأسماء الصحيحة
    get_color='[200, 30, 0, 160]',
    get_radius=50,
    pickable=True
)

view_state = pdk.ViewState(
    latitude=df['latitude'].mean(),
    longitude=df['longitude'].mean(),
    zoom=10,
    pitch=0,
)


r = pdk.Deck(layers=[point_layer], initial_view_state=view_state, tooltip={"text": "Lat: {latitude}, Lon: {longitude}"})
r.show()
```

```python
heatmap_layer = pdk.Layer(
    "HeatmapLayer",
    data=df,
    get_position='[longitude, latitude]',
    aggregation=pdk.types.String("MEAN"),
    get_weight=1,
)

view_state = pdk.ViewState(
    latitude=df['latitude'].mean(),
    longitude=df['longitude'].mean(),
    zoom=10,
    pitch=45,
)

r = pdk.Deck(layers=[heatmap_layer], initial_view_state=view_state)
r.show()
```
